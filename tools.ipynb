# Trimesh, convert .ply to .obj

```python
import trimesh
import os

# 1. Load your mesh (replace 'input_mesh.obj' with your file path)
# Trimesh supports formats like STL, OBJ, PLY, GLTF, etc.
input_path = '/home/frozzzen/Downloads/Milo_meshes/hotdog_50000.ply'
mesh = trimesh.load_mesh(input_path)

# 2. Define the output file path and desired format
output_path = '/home/frozzzen/Downloads/Milo_meshes/hotdog_50000_trimesh.obj' # Example: exporting as an STL file

# 3. Export the mesh
try:
    mesh.export(output_path)
    print(f"Mesh successfully exported to {os.path.abspath(output_path)}")
except Exception as e:
    print(f"An error occurred during export: {e}")
```
